```python
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
def combination(n, k):
    if(0 <= k <= n):
        return (math.factorial(n)/(math.factorial(n-k)*math.factorial(k)))
    else:
        return 0
```

```python
def pmf(l, k):
    return math.pow(l, k)/(math.factorial(k))*math.pow(math.e,-l)
```

```python
# Set seaborn std cofiguration to plots
sns.set()

x = list(range(11))
lambda_1 = [pmf(1, i) for i in x]
lambda_4 = [pmf(4, i) for i in x]
lambda_10 = [pmf(10, i) for i in x]

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(20,14))
ax1.bar(x,lambda_1,label='lambda = 1')
ax1.legend(loc=0)
ax2.bar(x,lambda_4,label='lambda = 4')
ax2.legend(loc=0)
ax3.bar(x,lambda_10,label='lambda = 10')
ax3.legend(loc=0)
ax4.set_visible(False)

fig.text(0.5, 0.9, 'PMF', horizontalalignment='center', color='black', weight='bold', size=20)
plt.show()
```

```python
def cdf(l, k):
    k_ = math.floor(k)
    euler_lambda = math.pow(math.e,-l)
    s = [math.pow(l,i)/(math.factorial(i)) for i in range(k_)]
    return euler_lambda*sum(s)
```

```python
# Set seaborn std cofiguration to plots
sns.set()

x = list(range(11))
lambda_1 = [cdf(1, i) for i in x]
lambda_4 = [cdf(4, i) for i in x]
lambda_10 = [cdf(10, i) for i in x]

fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2, 2, figsize=(20,14))
ax1.bar(x,lambda_1,label='lambda = 1')
ax1.legend(loc=0)
ax2.bar(x,lambda_4,label='lambda = 4')
ax2.legend(loc=0)
ax3.bar(x,lambda_10,label='lambda = 10')
ax3.legend(loc=0)
ax4.set_visible(False)

fig.text(0.5, 0.9, 'CDF', horizontalalignment='center', color='black', weight='bold', size=20)
plt.show()
```

```python
def Binomial(n, p, k):
    if(0 <= k <= n):
        return math.pow(p,k)*math.pow((1-p),(n-k))*combination(n, k)
    else:
        return 0
```

```python
# Set seaborn std cofiguration to plots
sns.set()

fig,ax = plt.subplots()

axis_x = list(range(11))
X = [Binomial(20, 0.05, i) for i in axis_x]
Y = [pmf(20*0.05,i) for i in axis_x]                      




ax.scatter(axis_x,X,color='red',label='X~binomial(20,0.05)',linewidths=5)
ax.scatter(axis_x,Y,label='Y~poisson(20*0.05)',linewidths=5)

plt.legend(loc=0)

fig.text(0.5, 0.9, 'Aproximação por Poisson', horizontalalignment='center', color='black', weight='bold', size=20)
plt.show()
```

```python
def hypergeom(k , n, d, m):
    return combination(d, k)*combination(n-d, m-k)/combination(n, m)
```

```python
# Set seaborn std cofiguration to plots
sns.set()

fig,ax = plt.subplots()

axis_x = list(range(21))
X = [hypergeom(i, 100, 5, 4) for i in axis_x]

ax.scatter(axis_x,X,label='X~hypergeom(k,100,5,4)',linewidths=5)

plt.legend(loc=0)

fig.text(0.5, 0.9, 'Distribuição hipergeométrica - pmf', horizontalalignment='center', color='black', weight='bold', size=20)
plt.show()
```

```python
def fda(k , n, d, m):
    return combination(m, k+1)*combination(n-m, d-k-1)/combination(n, d)
```

```python
# Set seaborn std cofiguration to plots
sns.set()

fig,ax = plt.subplots()

axis_x = list(range(21))
X = [1-fda(i, 100, 5, 4) for i in axis_x]

ax.scatter(axis_x,X,label='X~hypergeom(k,100,5,4)',linewidths=5)

plt.legend(loc=4)

fig.text(0.5, 0.9, 'Distribuição hipergeométrica - cdf', horizontalalignment='center', color='black', weight='bold', size=20)
plt.show()
```
