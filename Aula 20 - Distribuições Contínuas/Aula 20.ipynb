```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import math
def expo(x, l):
    if(x >= 0):
        return l*math.pow(math.e,(-l*x))
    else:
        return 0
```

```python
X = list(range(501))
X = [i/100.0 for i in X]
Y = [expo(i, 1) for i in X]

sns.set()

plt.plot(X,Y)
plt.show()
```

```python
Y1 = [expo(i, 1) for i in X]
Y4 = [expo(i, 4) for i in X]
Y10 = [expo(i, 10) for i in X]

sns.set()

plt.plot(X,Y1,label='lambda = 1')
plt.plot(X,Y4,label='lambda = 4')
plt.plot(X,Y10, label='lambda = 10')
plt.legend(loc=0)
plt.show()
```
